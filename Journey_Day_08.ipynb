 * *Write a program usin functions to find greatest of three number*

```python
def find_greatest(a, b, c):
    if a >= b and a >= c:
        return a
    elif b >= a and b >= c:
        return b
    else:
        return c

# Example usage:
result = find_greatest(10, 25, 15)
print("The greatest number is:", result)  # Output: 25
```

*Write a python program using functions  to convert fahrenhite to celcius*

```python
def celsius_to_fahrenheit(celsius):
    fahrenheit = (celsius * 9/5) + 32
    return fahrenheit

# Example usage:
temperature_f = celsius_to_fahrenheit(37)
print("Temperature in Fahrenheit:", temperature_f)
```

**Write a recursive function to calculate the sum of first n natural numbers.**

```python
def recursive_sum(n):
    # Base case: the sum of the first 1 natural number is 1
    if n <= 1:
        return n
    else:
        # Recursive case: n + sum of (n-1) numbers
        return n + recursive_sum(n - 1)

# Example Usage
n = 5
print(f"The sum of first {n} natural numbers is: {recursive_sum(n)}")
```

**Write a python function to print first n lines of the following pattern:**
***

```python
def print_pattern(n):
    for i in range(n, 0, -1):
        print("*" * i)

# Example Usage
print_pattern(3)
```

** Write a python function which converts inches to cms.Note: $1 \text{ inch} = 2.54 \text{ cm}$**

```python
def inches_to_cms(inches):
    return inches * 2.54

# Example Usage
inches_val = 10
cms_val = inches_to_cms(inches_val)
print(f"{inches_val} inches is equal to {cms_val} cm")
```

**Write a python function to remove a given word from a list ad strip it at the same time**

```python
def remove_and_strip(word_list, target_word):
    cleaned_list = []
    for word in word_list:
        # Remove trailing/leading spaces from the current word
        stripped_word = word.strip()
        
        # Only keep the word if it doesn't match the target word
        if stripped_word != target_word:
            cleaned_list.append(stripped_word)
            
    return cleaned_list

# Example Usage
fruits = ["  apple  ", " banana", "orange ", "  apple"]
word_to_remove = "apple"

updated_list = remove_and_strip(fruits, word_to_remove)
print(updated_list) 
# Output: ['banana', 'orange']
```

**Write a python function to print multiplication table of a given number.**

```python
def print_table(num):
    for i in range(1, 11):
        print(f"{num} x {i} = {num * i}")

# Example Usage
print_table(7)
```
