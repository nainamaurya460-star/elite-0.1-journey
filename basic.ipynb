**Question 4**
School Fees Discount: Ek function banao jo total fees input le, aur usme se 500 rupees minus (kam) karke bachi hui fees return kare.

```python
def fees_discount(total_fees):
    discount=total_fees-500
    return discount
payable_amount=fees_discount(34400)
print(payable_amount)
```
