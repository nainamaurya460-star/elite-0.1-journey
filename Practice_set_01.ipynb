❓ Question 1: Swapping Numbers
**Problem:** Swap the values of two variables `a = 10` and `b = 20` so that `a` becomes 20 and `b` becomes 10 without losing data.

```python
#### 💡 Solution:

# Initial values
a = 10
b = 20

# Using a temporary variable (The Backup Glass method)
temp = a
a = b
b = temp

# Alternative Python Shortcut (Uncomment to use):
# a, b = b, a

# print(f"After Swapping -> a: {a}, b: {b}")
# Output: After Swapping -> a: 20, b: 10
# simple
a=10
b=20
temp=a
a=b
b=temp
print(b)
print(a)
```

❓ Question 2: Type Caster
Problem: Take age as input from the user using input(), add 5 years to it, and print the updated age cleanly.

```python
# Taking input and typecasting string to integer explicitly
age = int(input("Enter your age: "))

# Mathematical processing
new_age = age + 5

# Printing using f-string formatting
print(f"After five years, your age will become: {new_age}")
```

❓ Question 3: Leap Year Checker
Problem: Check whether a user-entered year is a leap year or a regular year, handling centurial years correctly.

```python
year = int(input("Enter any year: "))

# Rule: Divisible by 4 AND not by 100, OR directly divisible by 400
if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
    print(f"{year} is a Leap Year! 📅")
else:
    print(f"{year} is a Regular Year.")
```

❓ Question 5: Table Generator (For Loop)
Problem: Generate a structured mathematical multiplication table (from 1 to 10) for any number using a for loop.

```python
num = int(input("Enter a number for its multiplication table: "))

print(f"\n--- Table of {num} ---")
# range(1, 11) runs the loop from 1 up to 10
for i in range(1, 11):
    print(f"{num} x {i} = {num * i}")
```

❓ Question 6: Counting with While Loop
Problem: Print all Odd numbers starting from 1 up to 20 using a conditional while loop.

```python
print("--- Odd Numbers (1 to 20) ---")
i = 1  # Loop Initialization

while i <= 20:
    print(i)
    i += 2  # Increment by 2 to skip even numbers directly
```
